```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Télécharger manuellement depuis Kaggle :
# https://www.kaggle.com/datasets/datamunge/sign-language-mnist

# Supposons que vous avez téléchargé sign_mnist_train.csv et sign_mnist_test.csv
# Placez-les dans data/raw/

# Charger les données
train_df = pd.read_csv('data/raw/sign_mnist_train.csv')
test_df = pd.read_csv('data/raw/sign_mnist_test.csv')

print(f"Données d'entraînement : {train_df.shape}")
print(f"Données de test : {test_df.shape}")
print(f"\nClasses présentes : {sorted(train_df['label'].unique())}")

# Visualiser quelques exemples
fig, axes = plt.subplots(2, 5, figsize=(15, 6))
for i, ax in enumerate(axes.flat):
    img = train_df.iloc[i, 1:].values.reshape(28, 28)
    label = train_df.iloc[i, 0]
    ax.imshow(img, cmap='gray')
    ax.set_title(f"Classe {label}")
    ax.axis('off')
plt.tight_layout()
plt.savefig('results/plots/sample_data.png')
plt.show()
```
